```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal as mvn
```

```python
# Parâmetros da primeira população
mu1 = np.array([0,0])
cov1 = np.array([[1,0], [0,4]])

# Parâmetros da segunda população
mu2 = np.array([0,0])
cov2 = np.array([[4,0], [0,1]])
```

```python
# Definindo os valores de pi_0 e pi_1
p0 = 0.5
p1 = 1 - p0

# Definindo os valores de c0 e c1
c0 = 1
c1 = 1

# Criando pontos 2D para plotarmos as curvas de nível da densidade
x, y = np.meshgrid(np.linspace(-4,4), np.linspace(-4,4))
xy = np.column_stack([x.flat, y.flat])

# Calculando as densidades
z1 = mvn.pdf(xy, mu1, cov1).reshape(x.shape)
z2 = mvn.pdf(xy, mu2, cov2).reshape(x.shape)

# Definindo alguns parâmetros do plot
plt.figure(figsize=(6,6))
plt.title('Curvas de nível para X|Y=0 e X|Y=1 e a fronteira de decisão')
plt.xlabel('X1')
plt.ylabel('X2')
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Plotando as curvas de nível
# A curva preta é referente à X1 e a vermelha à X2
plt.contour(x, y, z1, levels=5, colors='black')
plt.contour(x, y, z2, levels=5, colors='red')

# Calculando a fronteira ótima de Bayes, com as observações feitas!
A = np.linalg.inv(cov1) - np.linalg.inv(cov2)
alpha = -2 * np.log((c1 * p0) / (c0 * p1))

# Calculando a fronteira.
# Note que np.sign retorna -1 caso z < 0, 0 se z = 0 e 1 se z > 0
z = np.sign(A[0,0] * x**2 + 2*A[0,1]*x*y + A[1,1]*y**2 + alpha)

# Exibindo a fronteira, onde roxo = 0 e amarelo = 1
plt.imshow(z, extent=[-5, 5, -5, 5], alpha=0.4)
plt.show()
```

## Item 01

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Definindo nossa amostra
X = np.array([[4,1], [2,3], [5,4], [1,0]])

# Padronizando as variáveis
X_padronizado = StandardScaler().fit_transform(X)

# Computando os 2 componentes principais
# Para o PCA implementado pelo scikit-learn nós devemos
# providenciar a matriz de dados e não a de covariância!
pca = PCA(n_components=2)
pca.fit(X_padronizado)

# Podemos acessar os componentes através de pca.components_
# e suas variâncias explicadas através de pca.explained_variance_

pca.components_
```

```python
np.round(pca.components_, 5)
```

```python
np.round(pca.explained_variance_, 5)
```

## Item 02

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo nossa amostra
X = np.array([[4,1], [2,3], [5,4], [1,0]])

# Definindo o primeiro componente principal
v = pca.components_[0]

# Definindo algumas configurações do gráfico
plt.style.use('ggplot')
plt.figure(figsize=(8,8))
plt.xlabel('X1')
plt.ylabel('X2')
plt.xlim(-1, 6)
plt.ylim(-1, 6)

# Plotando a amostra
plt.scatter(X[:,0], X[:,1], label='amostra', color='green')

# Plotando o primeiro componente principal
plt.quiver([0], [0], v[0], v[1], scale=1, label='primeiro componente principal')

plt.legend(loc='upper left')
plt.show()
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(X, columns=['X1', 'X2'])
df['projecao'] = np.dot(X, v)

df
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Definindo a matriz de rotação
radians = 30 * (np.pi / 180)
rot = np.array([
    [np.cos(radians), -np.sin(radians)],
    [np.sin(radians), np.cos(radians)]
])

# Definindo a amostra rotacionada
X_rot = (rot @ X.T).T

# Padronizando as variáveis
X_rot_padronizado = StandardScaler().fit_transform(X_rot)

# Computando os componentes
pca_rot = PCA(n_components=2)
pca_rot.fit(X_rot_padronizado)

pca_rot.components_
```

```python
np.round(pca_rot.components_, 5)
```

```python
np.round(pca_rot.explained_variance_, 5)
```
